```python
# Identifying  the distinct possible values within the columns Type 1 and Type 2.

Type1_d = display(list(pokemon['Type 1'].unique()))
Type2_d = display(list(pokemon['Type 2'].unique()))
```

```python
# Identifying the string "Mega" within the column "Name". Transfer it to the beginning of the rowÂ´s content


def clean(x):
    if 'Mega' in x:
        change= x.split('Mega')[1]
        name= 'Mega'+ change
        return name
    else:
        return x

pokemon['Name']=pokemon['Name'].apply(clean)

```

```python
# new column "A/D Ratio" =  Attack / Defense.


pokemon['A/D Ratio'] = pokemon['Attack']/pokemon['Defense']
pokemon.head()
```

```python
# Creating a new column called âCombo Typeâ, combining the content of the columns âType1â and âType2â, or only Type1Â´s, 
# in case Type2 is not having a valid content

pokemon['Type 2']=pokemon['Type 2'].fillna('')

def combo_type(row):    
    if row['Type 2'] == '':
        return row['Type 1']
    else:
        return row['Type 1']+ '-' + row['Type 2']
    
pokemon['Combo Type'] = pokemon.apply(combo_type, axis=1)
pokemon.head()
```

```python
# Identifying the top 5 pokemon in terms of A/D Ratio 


top_5 = pokemon.sort_values(by='A/D Ratio', ascending=False).head(5)
top_5
```

```python
# identifying the unique values within the column "Combo Type", among the top_5 Pokemon


Combo_Type_u = display(list(top_5['Combo Type'].unique()))
```

```python
all_means = top_5.groupby('Combo Type').agg('mean').reset_index()
print(all_means)
# all_means[all_means['Combo Type'].isin(Combo_Type_u)]
```
